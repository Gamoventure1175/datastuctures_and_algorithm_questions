# Bubble Sort

## How it works?

For a given unsorted list **[d, c, b, a]**, the idea is as follows:
1. First pass -> [c, b, a, d]
2. Second pass -> [b, a, c, d]
3. Third pass -> [a, b, c, d]


Basically, for an array/list of *n* elements, we would be doing *n-1* passes
Also, each pass pushes the kth largest element to the kth largest position in the array which in simpler terms is:
* *Each pass will put the largest element at the end*

So, we won't have to check for/compare the largest elements at every inner loop iteration, which in simpler terms is:
* *Inner loop will shrink*

So, in pseudocode format:
* For pass in range 1 ... n-1
    * For i in range 1 ... n-pass - 1
        * Compare arr[i] > arr[i+1]:
            swap arr[i], arr[i+1]

```python
def bubbleSort(nums: list[int]):
    n = len(nums)
    for pass_number in range(0, n-1):
        swapped = False
        for i in range(0, n-pass_number-1):
            if nums[i] > nums[i+1]:
                nums[i], nums[i+1] = nums[i+1], nums[i]
                swapped = True
        if not swapped:
            return nums
    
    return nums
```

```python
example = [5, 2, 3 ,1]
print(bubbleSort(example))
```
