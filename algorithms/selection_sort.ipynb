# Selection Sort

## How it works? 

It is the opposite in terms of working compared to Bubble sort.

For a given list [5, 4, 1, 2, 3], selection sort would work in the following manner:
1. first pass -> [1, 4, 5, 2, 3]
2. second pass -> [1, 2, 5, 4, 3]
3. third pass -> [1, 2, 3, 4, 5]

So, with every pass, you move the smallest (or the largest) element to the first starting positions

>Also, compared to bubble sort algorithm, you don't swap positions between elements for every pass.
>Instead, with every pass, only the smallest element is swapped with the current starting element  

```python
def selection_sort(nums: list[int]):
    for pass_count in range(len(nums)-1):
        min_index = pass_count
        
        for i in range(pass_count+1, len(nums)):
            if nums[i] < nums[min_index]:
                min_index = i
        
        if min_index != pass_count:                
            nums[pass_count], nums[min_index] = nums[min_index], nums[pass_count]
        
    return nums                 
```

```python
example = [5, 4, 2, 1, 1, 10, 10, 5]
print(selection_sort(nums=example))
```
