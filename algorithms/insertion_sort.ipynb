# Insertion Sort

## How it works? 


It works like the analogy of sorting cards held in a hand. It divides the array into sorted and unsorted regions and then selects the first element from the unsorted region and inserts it into its right position in the sorted region by shifting the elements in the sorted region.

For example, sorting array [3, 4, 5, 1, 2] would something like this:
1. First pass -> [3, 4, 5, 1, 2]
2. Second pass -> [3, 4, 5, 1, 2]
3. Third pass -> [1, 3, 4, 5, 2]
4. Fourth pass -> [1, 2, 3, 4, 5]

Also, the outer loop would run for ***1 ... n-1*** positions whilst the inner loop would run in the opposite direction (backwards) of the outerloop for ***0 ... i-1***

> It divides the array into sorted and unsorted regions and only operates on the sorted region

Shifting and swapping between the elements in the sorted region would both result in the same correctness. However, shifting would help benefit the performance of the algorithm as it would not create unnecessary swapping everytime.

- [x] Best for scenarios where the array is almost sorted

```python
def insertion_sort(nums: list[int]):
    n = len(nums)
    for i in range(1, n):
        key = nums[i]
        j = i - 1
        
        while j >= 0 and nums[j] > key:
            nums[j+1] = nums[j]
            j -= 1 
        nums[j+1] = key

    return nums
```

```python
example = [3, 4, 2, 6]
print(insertion_sort(example))
```
